```python
train_data["trans_date_trans_time"] = pd.to_datetime(train_data["trans_date_trans_time"])
train_data["dob"] = pd.to_datetime(train_data["dob"])
```

```python
train_data1 = train_data.drop(columns=['Unnamed: 0','cc_num','first', 'last', 'street', 'city', 'state', 'zip', 'dob', 'trans_num','trans_date_trans_time'], axis=1)
```

```python
train_data1.columns
```

```python
train_data1.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
train_data1["merchant"] = encoder.fit_transform(train_data1["merchant"])
train_data1["category"] = encoder.fit_transform(train_data1["category"])
train_data1["gender"] = encoder.fit_transform(train_data1["gender"])
train_data1["job"] = encoder.fit_transform(train_data1["job"])
```

```python
train_data1
```

```python
exit_counts = train_data["is_fraud"].value_counts()
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.pie(exit_counts, labels=["No", "YES"], autopct="%0.0f%%")
plt.title("is_fraud Count")
plt.tight_layout()
plt.show()
```

```python
X = train_data1.drop(columns=["is_fraud"], inplace = False)
y = train_data1["is_fraud"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
model.score(X,y)
```
